```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import seaborn as sns
import scipy.stats 
from scipy.stats import pearsonr
# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_df = pd.merge(study_results, mouse_metadata, on='Mouse ID', how='left')
# Display the data table for preview
combined_df.head()
```

```python
# Checking the number of mice.
mice_count=combined_df['Mouse ID'].value_counts()
print(len(mice_count))
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice_id = combined_df[combined_df.duplicated(subset=["Mouse ID","Timepoint"], keep=False)]['Mouse ID'].unique()
duplicate_mice_id
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mice_data = combined_df[combined_df["Mouse ID"]=='g989']
duplicate_mice_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
drop_duplicate = combined_df[~combined_df['Mouse ID'].isin(duplicate_mice_id)]
drop_duplicate.head()
```

```python
# Checking the number of mice in the clean DataFrame.
unique_miceids = drop_duplicate['Mouse ID'].nunique()
print(unique_miceids)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
tumor_summary_stats = drop_duplicate.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])
tumor_summary_stats.columns= ['Mean Tumor Volume', 'Median Tumor Volume','Tumor Volume Variance', 'Tumor Volume Std. Dev','Tumor Volume Std. Err.']
tumor_summary_stats
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
tumor_stats = drop_duplicate.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])
tumor_stats
```

## Quartiles, Outliers and Boxplots

```python
# Assuming you have a DataFrame named 'data' containing the tumor volume data and treatment regimen information

# Get the last (greatest) timepoint for each mouse
max_timepoint_df = combined_df.groupby('Mouse ID')['Timepoint'].max().reset_index()

# Merge the max_timepoint_df with the original DataFrame to get the tumor volume at the last timepoint
final_tumor_volume_df = pd.merge(max_timepoint_df, combined_df, on=['Mouse ID', 'Timepoint'], how='left')

# Filter the DataFrame to include only the four specified treatment regimens
treatment_regimens = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
final_tumor_volume_df = final_tumor_volume_df[final_tumor_volume_df['Drug Regimen'].isin(treatment_regimens)]

# Display the DataFrame with the final tumor volume for each mouse across the specified treatment regimens
final_tumor_volume_df
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse


# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
greatest_timepoint = final_tumor_volume_df.groupby('Mouse ID')['Timepoint'].max().reset_index()
tumor_volume = pd.merge(greatest_timepoint,final_tumor_volume_df, on =['Mouse ID','Timepoint'], how='inner')
pd.merge(max_timepoint_df, combined_df, on=['Mouse ID', 'Timepoint'], how='inner')
med_used = ['Capomulin', 'Remicane','Infubinol', 'Ceftamin']
tumor_volume = tumor_volume[tumor_volume['Drug Regimen'].isin(med_used)]
tumor_volume
```

```python
# Assuming you have a DataFrame named 'data' containing the tumor volume data for each treatment group

# Group the data by the 'Drug Regimen' column
grouped_data = final_tumor_volume_df.groupby('Drug Regimen')

# Iterate over each treatment group to find potential outliers
for group_name, group_data in grouped_data:
    quartiles = group_data['Tumor Volume (mm3)'].quantile([0.25, 0.75])
    lower_quartile = quartiles[0.25]
    upper_quartile = quartiles[0.75]
    iqr = upper_quartile - lower_quartile
    
    lower_bound = lower_quartile - 1.5 * iqr
    upper_bound = upper_quartile + 1.5 * iqr
    
    potential_outliers = group_data[(group_data['Tumor Volume (mm3)'] < lower_bound) | (group_data['Tumor Volume (mm3)'] > upper_bound)]['Tumor Volume (mm3)']
    
    print(f"{group_name}'s potential outliers: {potential_outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
grouped_data = final_tumor_volume_df.groupby('Drug Regimen')
treatments=[]
tumor_volumes=[]
for group_name, group_data in grouped_data:
    treatments.append(group_name)
    tumor_volumes.append(group_data['Tumor Volume (mm3)'])
# Create a boxplot
plt.figure(figsize=(8,6))
plt.boxplot(tumor_volumes, labels=treatments)
plt.title('Tumor Volume Distribution by Treatment Group')
plt.xlabel('Treatment Group')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
capomulin_data = combined_df[combined_df['Drug Regimen'] == 'Capomulin']
mouse_id = 'l509'
mouse_data = capomulin_data[capomulin_data['Mouse ID'] == mouse_id]
plt.figure(figsize=(8, 6))
plt.plot(mouse_data['Timepoint'], mouse_data['Tumor Volume (mm3)'], marker='o', linestyle='-', color='b')
plt.title('Capomulin Treatment of mouse l509')
plt.xlabel('Timepoint')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
weight_avg_tumor = capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean().reset_index()
merged_data = pd.merge(weight_avg_tumor, mouse_metadata, on = 'Mouse ID')
plt.figure(figsize=(7,5))
plt.scatter(merged_data['Weight (g)'], merged_data['Tumor Volume (mm3)'], color='b', alpha=1)
plt.xlabel('Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
plt.show
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
from sklearn.linear_model import LinearRegression
weight_avg_tumor = capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean().reset_index()
merged_data = pd.merge(weight_avg_tumor, mouse_metadata, on= 'Mouse ID')
correlation,_ = pearsonr(merged_data['Weight (g)'], merged_data['Tumor Volume (mm3)'])
correlation_rounded = round(correlation, 2)
print(f"The correlation between mouse weight and the average tumor volume is :{correlation_rounded}")
X = merged_data[['Weight (g)']]
y = merged_data['Tumor Volume (mm3)']
model = LinearRegression()
model.fit(X,y)
slope = model.coef_[0]
intercept = model.intercept_
plt.figure(figsize= (7,5))
plt.scatter(merged_data['Weight (g)'], merged_data['Tumor Volume (mm3)'], color= 'b', alpha=1)
plt.plot(X, model.predict(X), color='r', linestyle="-", linewidth=2)
plt.xlabel('Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```
